```python
import pandas as pd
```

```python
df = pd.read_csv("gdp to adoption.csv")
df
```

```python
# making the first row as the first column
df = df.transpose()
```

```python
# making the first row header
df = df.rename(columns=df.iloc[0])
```

```python
# dropping the first row now that it was copied to header
df = df.drop(df.index[0])
```

```python
# making the gdp values look small by dividing a billion
df['gdp(billion)'] = df['gdp'] / 1000000000
```

```python
# and round it
decimals = 2
df['gdp(billion)'] = df['gdp(billion)'].apply(lambda x: round(x, decimals))
```

```python
# dropping the original gdp column
df = df.drop('gdp', axis=1)
```

```python
df
```

```python
df.to_csv("gdp comparison.csv")
```
